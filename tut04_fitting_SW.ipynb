```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sesws_lib as sesws




        
```

```python
dir = 'Data'
fname = 'sample001'
offx, offy = 0, 0
tx, ty = 0,0

ex, ey = -1,-1
p = 0.5
cutx,cuty = 0,0
z_pos_num = 211
nonoff = 1

# Call library
lib = sesws.Shape_evaluation(dir,fname,offx,offy,tx,ty,ex,ey)

# pick up one structure
m_cut = lib.Find_one_structure(p,cuty,cutx)
# If you find strange cut m please change the initial values of cut x and cut y

# depth of used structure [mm]
z_pos_arr = np.linspace(np.max(m_cut),np.min(m_cut),z_pos_num)

X,Y,Z = lib.Output_fitted_parameters(m_cut,z_pos_arr,p, nonoff,sname = 'Sample001')

lib.Datwrite('Sample001_fit',Z,z_res = 210)
```
