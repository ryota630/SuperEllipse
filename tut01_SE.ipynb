# Import lib

```python
import numpy as np
import matplotlib.pyplot as plt
import sesws_lib as sesws
```

# Plot super ellipse

## Symmetric case

```python
nx_arr = np.array([0.5,1.0,1.5,2.0,2.5,3.0])
ny_arr = np.array([0.5,1.0,1.5,2.0,2.5,3.0])
rx_arr = np.ones(len(nx_arr))
ry_arr = np.ones(len(ny_arr))
sname = 'tut01-1'

lib1 = sesws.Gen_SuperEllipse()
x_arr,y_arr = lib1.Contour_SuperEllipse_arr(rx_arr,ry_arr,nx_arr,ny_arr)
lib1.Plot_Contour_SuperEllipse(x_arr,y_arr,nx_arr,ny_arr,sname)
```

## Asymmetric radius in $x$ and $y$

```python
nx_arr = np.array([0.5,1.0,1.5,2.0,2.5,3.0])
ny_arr = np.array([0.5,1.0,1.5,2.0,2.5,3.0])
rx_arr = np.ones(len(nx_arr))*1.2
ry_arr = np.ones(len(ny_arr))
sname = 'tut01-2'

lib1 = sesws.Gen_SuperEllipse()
x_arr,y_arr = lib1.Contour_SuperEllipse_arr(rx_arr,ry_arr,nx_arr,ny_arr)
lib1.Plot_Contour_SuperEllipse(x_arr,y_arr,nx_arr,ny_arr,sname)
```
