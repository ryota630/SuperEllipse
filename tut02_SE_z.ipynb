# Import lib

```python
import numpy as np
import matplotlib.pyplot as plt
import sesws_lib as sesws
```

# Sample code

```python
lib2 = sesws.Gen_SuperEllipse_z()
z_num = 101       # number of data in depth
p = 0.5           # pitch in mm
r_top = 0.075     # radius for the top of sws in mm
d_tot = 2         # total depth in mm

z_arr = np.linspace(0,d_tot,z_num)               # depth array
rx_arr = np.linspace(r_top,p/2*np.sqrt(2),z_num) # radius array in x
ry_arr = np.linspace(r_top,p/2*np.sqrt(2),z_num) # radius array in y
nx_arr = np.linspace(5.0,2.0,z_num)              # index of super ellipse in x
ny_arr = np.linspace(5.0,2.0,z_num)              # index of super ellipse in y

# contour along z axis
resx,resy = lib2.SuperEllipse_zdep(z_num,rx_arr,ry_arr,nx_arr,ny_arr,p) 

# Plot
lib2.Plot_Contour_SuperEllipse_z(z_arr,rx_arr,ry_arr,resx,resy,nx_arr,ny_arr,'tut02')
```
